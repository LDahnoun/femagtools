### Run AMELA in femagtools ###

```python
import logging 
import matplotlib.pyplot as plt 
from femagtools import amela
```

```python
FORMAT = "%(asctime)s %(name)s: %(message)s"
logging.basicConfig(level=logging.INFO, format=FORMAT)
```

If the nc file and the AMELA are in the same directory, only the workdir is needed. Otherwise, both the workdir (nc file) and amela_dir (amela executable) are needed. 

```python
workdir = '../../../amela_interface'
# test is the name of the nc file (femag model)
amela = amela.Amela(workdir, dict(name='test'))
r = amela()
```

```python
import matplotlib.pyplot as plt 
plt.plot(r['pm_data_se38']['loss_data'][:, 0]*1e3, r['pm_data_se38']['loss_data'][:, 1])
plt.grid()
plt.xlabel('Time / ms')
plt.ylabel('Loss / W')
```
